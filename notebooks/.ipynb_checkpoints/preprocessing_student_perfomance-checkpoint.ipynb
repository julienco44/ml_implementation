```python
def evaluate(model, X_train, y_train, X_test, y_test):
    """Evaluate model performance."""
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    return {
        'train_mse': mean_squared_error(y_train, y_train_pred),
        'train_r2': r2_score(y_train, y_train_pred),
        'test_mse': mean_squared_error(y_test, y_test_pred),
        'test_rmse': np.sqrt(mean_squared_error(y_test, y_test_pred)),
        'test_mae': mean_absolute_error(y_test, y_test_pred),
        'test_r2': r2_score(y_test, y_test_pred),
        'y_test_pred': y_test_pred
    }

def print_results(results, name):
    """Print evaluation results."""
    print(f"\n{name}")
    print("-" * 45)
    print(f"  Train MSE: {results['train_mse']:>12,.0f}  RÂ²: {results['train_r2']:.4f}")
    print(f"  Test  MSE: {results['test_mse']:>12,.0f}  RÂ²: {results['test_r2']:.4f}")
    print(f"  Test RMSE: {results['test_rmse']:>12,.2f}  MAE: {results['test_mae']:.2f}")
```

```python
RF_PARAMS = {
    'n_estimators': 50,
    'max_depth': 10,
    'min_samples_split': 10,
    'min_samples_leaf': 5,
    'max_features': "sqrt",
    'bootstrap': True,
    'random_state': 42
}

print(f"\nRandom Forest Hyperparameters: {RF_PARAMS}")
print("="*60)

start = time.time()
our_rf = RandomForestRegressorScratch(**RF_PARAMS)
our_rf.fit(X_train, y_train)
rf_time = time.time() - start

rf_results = evaluate(our_rf, X_train, y_train, X_test, y_test)
rf_results['train_time'] = rf_time
print_results(rf_results, 'Our RandomForest')
print(f"  Train time: {rf_time:.3f}s")
```

```python
sk_rf = RandomForestRegressor(
    n_estimators=50,
    max_depth=10,
    min_samples_split=10,
    min_samples_leaf=5,
    max_features="sqrt",
    bootstrap=True,
    random_state=42,
    n_jobs=-1
)

start = time.time()
sk_rf.fit(X_train, y_train)
sk_rf_time = time.time() - start

sk_rf_results = evaluate(sk_rf, X_train, y_train, X_test, y_test)
sk_rf_results['train_time'] = sk_rf_time
print_results(sk_rf_results, 'Sklearn RandomForest')
print(f"  Train time: {sk_rf_time:.3f}s")
```

```python
n_estimators_values = [10, 30, 50, 100]
max_depth_values = [5, 10, 15]
max_features_values = ["sqrt", None]  # None = all features

rf_grid_results = []

for n_trees in n_estimators_values:
    for depth in max_depth_values:
        for mf in max_features_values:
            params = {
                'n_estimators': n_trees,
                'max_depth': depth,
                'min_samples_split': 10,
                'min_samples_leaf': 5,
                'max_features': mf,
                'bootstrap': True,
                'random_state': 42
            }
            
            model = RandomForestRegressorScratch(**params)
            model.fit(X_train, y_train)
            res = evaluate(model, X_train, y_train, X_test, y_test)
            
            rf_grid_results.append({
                'n_estimators': n_trees,
                'max_depth': depth,
                'max_features': mf,
                'train_mse': res['train_mse'],
                'train_r2': res['train_r2'],
                'test_mse': res['test_mse'],
                'test_rmse': res['test_rmse'],
                'test_mae': res['test_mae'],
                'test_r2': res['test_r2']
            })

df_rf_grid = pd.DataFrame(rf_grid_results)
df_rf_grid
```
